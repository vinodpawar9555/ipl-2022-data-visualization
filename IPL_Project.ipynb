```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('IPL.csv')
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
# Which team won the most matches
```

```python
match_wins = df['match_winner'].value_counts()
sns.barplot(y=match_wins.index, x=match_wins.values)
plt.title('Teams with Most Matches Won')
```

```python
# Toss decision trend
```

```python
sns.countplot(x = df['toss_decision'])
```

```python
# Toss Winner vs Match Winner
```

```python
count = df[df['toss_winner'] == df['match_winner']]['match_id'].count()
percentage = (count/len(df))*100
percentage.round(2)
```

```python
# How do teams win?
```

```python
sns.countplot(x = df['won_by'])
```

```python
# Most player of the match award
```

```python
count = df['player_of_the_match'].value_counts().head(10)
count
```

```python
sns.barplot(x = count.values, y = count.index, palette = 'rainbow')
plt.title('Most Player of the Match Awards')
```

```python
# 2 Top Scorers
```

```python
df.head()
```

```python
top_scorer = df.groupby('top_scorer')['highscore'].sum().sort_values(ascending = False).head(2)
top_scorer
```

```python
top_scorer.plot(kind = 'bar')
```

```python
# Best Bowler
```

```python
df.head()
```

```python
df['highest_wickets'] = df['best_bowling_figure'].apply(lambda x: x.split('--')[0])
df['highest_wickets'] = df['highest_wickets'].astype(int)
top_bowlers = df.groupby('best_bowling')['highest_wickets'].sum().sort_values(ascending = False).head(10)
top_bowlers
```

```python
top_bowlers.plot(kind = 'barh')
```
